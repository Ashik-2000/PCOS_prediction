# Importing Library

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

# Loading data

```python
PCOS_inf = pd.read_csv("PCOS_infertility.csv")
PCOS_woinf = pd.read_csv("PCOS_W_infertility.csv")
```

```python
PCOS_inf.info()
# Col: AMH(ng/mL) cotains numerical values as strings
```

```python
PCOS_woinf.info()
# Null values:
# N460(mrriage status)
# AH158(FastFood)
# col: AMH(ng/mL) cotains numerical values as strings
```

# Preprocessing

```python
# Merging two dataset
data = pd.merge(PCOS_woinf,PCOS_inf, on='Patient File No.', suffixes=('','_y'),how='left')

# Eliminating repeated column after merging
data =data.drop(['Unnamed: 42', 'Sl. No_y', 'PCOS (Y/N)_y', 'AMH(ng/mL)_y'], axis=1)

data.head()
```

```python
data.info()
```

```python
# Numerical data with dtype objective
data["AMH(ng/mL)"].head()
```

Here in the data set, one column has numerical values in string format 
and Two column has null values: Marraige Status (Yrs), Fast food (Y/N)

```python
# Converting string to numerical format
data["AMH(ng/mL)"] = pd.to_numeric(data["AMH(ng/mL)"], errors='coerce')

#Filling the null values
data['Marraige Status (Yrs)'].fillna(data['Marraige Status (Yrs)'].median(),inplace=True)
data['Fast food (Y/N)'].fillna(data['Fast food (Y/N)'].median(),inplace=True)
data['AMH(ng/mL)'].fillna(data['AMH(ng/mL)'].median(),inplace=True)
```
